En python no existen tipos específicos para las pilas y colas, se implementan mediante listas

* Una pila es una lista en la que se extraen elementos por el final y se extraen por el final
* Una cola es una lista en la que se extraen elementos por el final y se extraen por el inicio

```python
pila = []

# Apilar elementos
pila.append("Plato 1")
pila.append("Plato 2")
pila.append("Plato 3")

# Quitar el último (Plato 3)
print(pila)
ultimo = pila.pop()
print(f"Salió: {ultimo}")
print(pila)
```

```python
# Para las colas usamos el tipo deque que es una lista doblemente enlazada
from collections import deque

cola = deque()

# Llegan personas a la fila
cola.append("Persona A")
cola.append("Persona B")
cola.append("Persona C")

# Atender a la primera (Persona A)
print(cola)
primero = cola.popleft()
print(f"Atendiendo a: {primero}")
print(cola)
```
