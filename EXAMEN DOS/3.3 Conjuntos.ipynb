Los conjuntos son similares a las listas pero:


*   No están ordenados
*   No se pueden repetir elementos
* En lugar de los símbolos [] usamos {}



```python
# Declaración
conjunto = {'rojo', 'amarillo', 'verde'}
print(conjunto) # No tiene que mostrarse en el orden que se inicializó
```

```python
# Los elementos no se pueden repetir
conjunto = {'verde', 'rojo', 'amarillo', 'verde'}
print(conjunto)
```

```python
# No se puede acceder a un elemento por posición
print(conjunto[0])
```

```python
# Buscando elementos
conjunto = {'rojo', 'amarillo', 'verde'}
print('rojo' in conjunto)
print('azul' in conjunto)
```

```python
# Recorrido
conjunto = {'rojo', 'amarillo', 'verde'}
for i in conjunto:
  print (i)
```

```python
# Hay 3 operadores del ajebra de conjuntos:
# - unión(|): la "suma de los dos conjuntos"
# - intersección(&): los elementos comunes de los dos conjuntos
# - Diferencia(-): los elementos que están en un conjunto pero no en el otro
c1 = {'rojo', 'amarillo', 'verde'}
c2 = {'rojo', 'azul', 'negro'}
print (c1 | c2)
print (c1 & c2)
print (c1 - c2)
print (c2 - c1)
```

```python
# Podemos convertir un conjunto en una lista
conjunto = {'rojo', 'amarillo', 'verde'}
lista = list(conjunto)
print(lista[0])
```

```python
# Eliminando duplicados de una lista
l1 = [1,2,2,4,7,5,4]
conjunto = set(l1)
l2 = list(conjunto)
print(l1)
print(l2)
```

```python
# Otros métodos útiles:
# .add(x):  	  Añade un elemento (si no existe).
# .remove(x):	  Elimina un elemento (da KeyError si no existe).
# .discard(x):	Elimina un elemento (NO da error si no existe).
# .pop():	      Elimina y te devuelve un elemento aleatorio.
conjunto = {'rojo', 'amarillo', 'verde'}
conjunto.add('negro')
print(conjunto)

conjunto.remove('verde')
print(conjunto)

conjunto.discard('negro')
conjunto.discard('blanco')
print(conjunto)

color = conjunto.pop()
print(color)
print(conjunto)
```
